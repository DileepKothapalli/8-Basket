```python

def get_unique(team_name,season='2022-23'):
    team = pd.read_csv(f'./teams/teams_{season}/{team_name}.csv')
    display(team.PLAYER_NAME.unique())
    

# get player stats 
def get_player_stats(player,team_name,season='2022-23'):
    team = pd.read_csv(f'./teams/teams_{season}/{team_name}.csv')
    player_stats =  team.loc[team['PLAYER_NAME'] == player ]
    player_stats.sort_values(by='FANTASY_PTS',ascending=False)
    player_stats_refined = player_stats.drop(['SEASON_ID','PLAYER_ID','TEAM_ID','GAME_ID','VIDEO_AVAILABLE','Unnamed: 0','Unnamed: 0.1'],axis=1)
    pd.set_option('display.max_row',None)
    display(player_stats_refined.describe())



# get fantasy points of team
def get_fantasy_points_of_team(team_name,season='2022-23'):
    team = pd.read_csv(f'./teams/teams_{season}/{team_name}.csv')
    player_fantasy_points = team[['PLAYER_NAME','FANTASY_PTS']]
    grouped = (team.groupby('PLAYER_NAME')['FANTASY_PTS'].describe())
    display(len(grouped))
    display(grouped.sort_values(by='mean',ascending=False))

# get fantasy points of player all seasons


def get_fantasy_points_of_player_all_seasons(player):
    player_all_seasons = pd.DataFrame()
    seasons =  []
    seasons_num = 10
    for i in range(seasons_num):
        first = 2022-i
        second= 23-i
        seasons.append(str(first)+'-'+str(second))

    games_df =[]

    for season in seasons:
        team = pd.read_csv(f'./nba-scraped-data/player_games/player_game_{season}.csv')
        player_stats = team.loc[team['PLAYER_NAME'] == player]
        player_all_seasons[season] = player_stats['FANTASY_PTS'].reset_index(drop=True)
    display(player_all_seasons.describe())


# get_fantasy_points_of_player_all_seasons('Giannis Antetokounmpo')
# get_fantasy_points_of_player_all_seasons('Giannis Antetokounmpo')
# get_fantasy_points_of_player_all_seasons('Stephen Curry')
# get_player_stats('Stephen Curry','GSW',season='2022-23')
# get_fantasy_points_of_team('MIL')
```
